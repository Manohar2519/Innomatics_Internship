## KNN Imputation on Train Data

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
dff = pd.DataFrame({'Health':df_train['Health Indicator'],'Upper_age':df_train['Upper_Age'],'Lower_age':df_train['Lower_Age'],'Reco_Policy_Cat':df_train['Reco_Policy_Cat']})  
mm = MinMaxScaler()
def find_category_mappings(dff, variable):
    return {k: i for i, k in enumerate(dff[variable].dropna().unique(), 0)}
def integer_encode(dff , variable, ordinal_mapping):
    dff[variable] = dff[variable].map(ordinal_mapping)
mappin = dict()
def imputation(dff1 , cols):
    dff = dff1.copy()
    #Encoding dict &amp; Removing nan    
    #mappin = dict()
    for variable in cols:
        mappings = find_category_mappings(dff, variable)
        mappin[variable] = mappings
    #Apply mapping
    for variable in cols:
        integer_encode(dff, variable, mappin[variable])  
    #Minmaxscaler and KNN imputation 
    sca = mm.fit_transform(dff)
    knn_imputer = KNNImputer()
    knn = knn_imputer.fit_transform(sca)
    dff.iloc[:,:] = mm.inverse_transform(knn)
    for i in dff.columns : 
        dff[i] = round(dff[i]).astype('int')
    #Inverse transform
    for i in cols:
        inv_map = {v: k for k, v in mappin[i].items()}
        dff[i] = dff[i].map(inv_map)
    return dff
#Imputation
a=imputation(dff,['Health'])
```
