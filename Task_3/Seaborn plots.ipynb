```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\V.MANOHAR\OneDrive\Documents\Manu\Notes\data\temp\aspiring_minds_employability_outcomes_2015.csv")
df.head()
```

### 1.lmplot

lmplot compares numeric variables

It is a bivariate plot

It contains Numerical variables 

```python
sns.lmplot(data=df, x='10percentage',y='Salary',hue='Degree')
```

Observation:- 

-> From this dataset 90% Students have more Salary 

-> Maximum students having equal salary between 0 to 1

### 2.relplot

Relplot shows the relationship between two variables with semantic mappings of subsets. 

It is a bivariate plot

It contains Categorical and Numerical variables

```python
sns.relplot(data=df, x='Degree',y='Salary',hue='12percentage')
```

Observation:-
 
-> Higest number of Students are from B.Tech/B.E

-> Higest percentage in 12th got Higest Salary

### 3.violinplot

Violin plots are a method of plotting numeric data .

It can be considered a combination of the box plot with a kernel density plot.

It is a bivariate plot.

It contains Categorical and Numerical variables

```python
sns.violinplot(data=df, x='Degree',y='Salary',hue='Gender')
```

Observation:-
    
-> Higest Male candidates are earning more salary in MCA and Least Male candidates are in M.Sc(Tech)

-> Higest Female candidates are earning more salary in B.Tech/B.E and least Female candidates are in M.sc(Tech)

### 4.boxenplot

boxenplot() Draw an enhanced box plot for larger datasets. 

This style of plot was originally named a “letter value” plot because it shows a large number of quantiles that are defined as “letter values”.

It is a bivariate plot.

It contains Categorical and Numerical variables

```python
sns.boxenplot(data=df, x='Gender',y='Salary',hue='Degree')
```

Observation:-

-> Male Candidates earning more salary than Female Candidates

-> Higest salary is in B.Tech/B.E and Least Salary is in M.Sc(Tech)

### 5.catplot

Catplot is a relatively new addition to Seaborn that simplifies plotting that involves categorical variables. 

It changed the older factor plot to catplot to make it more consistent with terminology in pandas and in seaborn.

It is a bivariate plot.

It contains Categorical and Numerical variables

```python
sns.catplot(data=df, x='Gender',y='10percentage',hue='Degree')
```

Observations:-

-> Most of the B.Tech/B.E Students got Higest percentage in 10th

-> M.Sc Students got least percentage in 10th

### 6.pointplot

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables.

It is a bivariate plot.

It contains Categorical and Numerical variables

```python
sns.pointplot(data=df, x='Gender',y='Salary',hue='Degree')
```

Observation:-

On the basis of Degree Comparing Salary between Male and Female 

### 7.regplot

regplot() performs a simple linear regression model fit and plot.

lmplot() combines regplot() and FacetGrid

It is a bivariate plot.

It contains only Numerical variables

```python
sns.regplot(data=df, x='10percentage',y='Salary')
```

Observation:-
    
-> By using regplot() students got Higest salary at 55% and 90%

### 8.kdeplot

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.

It is a Univariate plot

It contains only Numerical variable

```python
sns.kdeplot(data=df, x='Salary',hue='Gender')
```

Observation:-

Male has the higest density than female

### 9.ecdfplot

An ECDF represents the proportion or count of observations falling below each unique value in a dataset. 

It is a Univariate plot

It contains Categorical and Numerical variables

```python
sns.ecdfplot(data=df, x='10percentage',hue='Degree')
```

Observation:-

From this data M.Sc(Tech) students have equal percentage in 10th

### 10.rugplot

It is used to visualize the distribution of the data.

It is a bivariate plot

It contains only Numerical variables

```python
sns.rugplot(data=df, x='10percentage',y='Salary',hue='Gender')
```

Observation:-
    
rugplot() visualising that on the basis of 10th percentage, Male candidates having more Salary than female candidates 

### 11.factorplot

factorplot() method is used to draw a categorical plot onto a FacetGrid.

It is a bivariate plot

It contains Categorical and Numerical variables

```python
sns.factorplot(data=df, x='CollegeTier',y='Salary',hue='Gender')
```

Observation:-
   
On the basis of CollegeTier Male has the highest factor in Salary than female

### 12.Residplot

This method will regress y on x and then draw a scatter plot of the residuals.

It is a bivariate plot

It contains only Numerical variables

```python
sns.residplot(data=df, x='CollegeTier',y='10percentage',)
```

Observation:-

2.0 College tier having more percentage in 10th
